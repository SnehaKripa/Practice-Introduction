```python
#1
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

# Create dummy data
data = {
    'ZN': np.random.rand(100),
    'INDUS': np.random.rand(100) * 2,
}

housing_data = pd.DataFrame(data)
plt.style.use('seaborn-darkgrid')

plt.figure(figsize=(12, 8))

plt.plot(housing_data['ZN'], color='green', label='ZN', linestyle='-')

plt.plot(housing_data['INDUS'], color='blue', label='INDUS', linestyle='--')

plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

plt.title('Line Plot of ZN and INDUS')

plt.legend()

plt.show()
```

```python
data = {
    'col1': np.random.randint(1, 10, 5),
    'col2': np.random.randint(1, 10, 5),
}
index = ['A', 'B', 'C', 'D', 'E']

df = pd.DataFrame(data, index=index)

plt.figure(figsize=(10, 6))
plt.bar(df.index, df['col1'], label='col1', color='b', alpha=0.6)
plt.bar(df.index, df['col2'], label='col2', color='g', alpha=0.6, bottom=df['col1'])

plt.legend(loc='lower left')

plt.title('Bar Chart of col1 and col2')

plt.show()

plt.figure(figsize=(10, 6))
plt.barh(df.index, df['col1'], label='col1', color='b', alpha=0.6)
plt.barh(df.index, df['col2'], label='col2', color='g', alpha=0.6, left=df['col1'])


plt.legend(loc='upper right')

plt.title('Horizontal Bar Chart of col1 and col2')

plt.show()
```

```python
# Create dummy data for MEDV
np.random.seed(0)
medv_values = np.random.normal(22, 5, 100)  # Generate 100 random values

# Create a DataFrame with the dummy data
data = pd.DataFrame({'MEDV': medv_values})

# Set the number of bins to 20
num_bins = 20

# Create the histogram
plt.figure(figsize=(10, 6))
plt.hist(data['MEDV'], bins=num_bins, color='skyblue', edgecolor='black')

# Add labels and a title
plt.xlabel('Median Value of Owner-Occupied Homes (MEDV)')
plt.ylabel('Frequency')
plt.title('Histogram of MEDV (Dummy Data)')

# Show the histogram
plt.show()

```

```python
np.random.seed(0)
x = np.random.rand(100)  # Generate random data for the x-axis
y = 2 * x + 1 + 0.2 * np.random.randn(100)  # Create a positive correlation with some noise

# Calculate the correlation coefficient (just to confirm the correlation)
correlation_coefficient = np.corrcoef(x, y)[0, 1]

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, c='b', label=f'Correlation = {correlation_coefficient:.2f}')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Scatter Plot of Two Positively Correlated Variables')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python

np.random.seed(0)
x = np.random.rand(100)
y = -2 * x + 1 + 0.2 * np.random.randn(100)

correlation_coefficient = np.corrcoef(x, y)[0, 1]

plt.figure(figsize=(8, 6))
plt.scatter(x, y, c='r', label=f'Correlation = {correlation_coefficient:.2f}')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Scatter Plot of Two Negatively Correlated Variables')
plt.legend()


plt.grid(True)
plt.show()
```
